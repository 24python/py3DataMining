```python
import pandas as pd

import numpy as np

df = pd.read_csv('data/HR.csv')
df = df.dropna()
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
f = plt.figure()
f.add_subplot(1,3,1)
sns.distplot(df['satisfaction_level'].dropna(),bins=10,kde=False)
f.add_subplot(1,3,2)
sns.distplot(df['last_evaluation'],bins=10)
f.add_subplot(1,3,3)
sns.distplot(df['average_monthly_hours'],bins=10)
```

报错: ValueError: max must be larger than min in range parameter.

原因是我们的satisfaction_level中有两个空值

绘制出的直方图之外，还画出了一条分布曲线。

- 当kde是false时，这条线就没有了
- 当hist是false时，直方图也就没有了

可以看到图中的波谷，基本上可以说明工作时长是成两极分布的
