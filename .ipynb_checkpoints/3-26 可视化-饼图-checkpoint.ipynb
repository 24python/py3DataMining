```python
import pandas as pd

import numpy as np

df = pd.read_csv('data/HR.csv')
df = df.dropna()
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
# 使用seaborn改变样式
# sns.set_style(style="darkgrid")
sns.set_style(style="whitegrid")
```

```python
lbs = df['department'].value_counts().index
plt.pie(df['department'].value_counts(normalize=True),
        labels=lbs,autopct="%1.1f%%",colors=sns.color_palette("Reds"))
```

如果我们想把销售部门的信息和整体分隔开来强调该部门

```python

explodes=[0.1 if i=="sales" else 0 for i in lbs]
plt.pie(df['department'].value_counts(normalize=True),explode=explodes,
        labels=lbs,autopct="%1.1f%%",colors=sns.color_palette("Reds"))
```

```python
lbs = df['salary'].value_counts().index
explodes=[0.1 if i=="high" else 0 for i in lbs]
plt.pie(df['salary'].value_counts(normalize=True),explode=explodes,
        labels=lbs,autopct="%1.1f%%",colors=sns.color_palette("Reds"))
```
